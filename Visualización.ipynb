```python
import pandas as pd

datos = pd.read_csv("MUERTES COVID NL.csv")
```

```python
sexo = datos['SEXO']
mingr = datos['FECHA_INGRESO']
muerte = datos['FECHA_DEF']
edad = datos['EDAD']
diasint = datos['T_INT']
embarazo = datos['EMBARAZO']
datos.head()
```

```python
edad.describe()
```

```python
diasint.describe()
```

```python
#número de mujeres fallecidas
mujer = 0
for i in sexo:
    if i == "F":
        mujer = mujer + 1
print(mujer) 
```

```python
#número de hombres fallecidos
hombre = len(sexo) - mujer
print(hombre)
```

```python
embarazadas = 0
for i in embarazo:
    if i == "Si":
        embarazadas = embarazadas + 1
print(embarazadas)
```

```python
import seaborn as sns
sns.pairplot(datos[['FECHA_INGRESO','FECHA_DEF', 'T_INT']])
```

```python
EdadF=datos.EDAD[datos['SEXO']=="F"]
EdadM=datos.EDAD[datos['SEXO']=="M"]
DatosEdades=[EdadF,EdadM]
```

```python
box=plt.boxplot(DatosEdades,labels=["Mujeres","Hombres"],patch_artist=True)
colors = ['magenta', 'blue']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
 
plt.show()
```
